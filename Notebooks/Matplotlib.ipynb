# Markers
We can use keyboard argument `marker` to emphasize each point with a specified marker.

```python
import matplotlib.pyplot as  plt
import numpy as np

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="*") #marker="o" for topping as 0. 
plt.show()
```

We can always find marker reference and use them according to our needs. 

Format strings: `fmt` 
`marker/Line/color` syntax can be used in shortcut string notation parameter to specify the marker. 

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,9,1,10])
plt.plot(ypoints, "o--b")
plt.show()
```

The marker value can be anything from Marker Reference and color value as well. 

```python
# Marker size: ms to set the size of markers. 

import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="o",ms=15)
plt.show()
```

```python
# Marker egde color by using mec, to set egde of marker.

import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([6,10,12,14])
plt.plot(ypoints,marker="o",ms=20,mec="g")
plt.show()
```

```python
# To set the markerface color, we use mfc to set color inside the edge of the markers.

import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([4,12,16,24])
plt.plot(ypoints,marker="*",ms=20,mfc="k")
plt.show()
```

## Pie-chart

- To create pie chart with pyplot, we can use `pie()` . 
- The start angle parameter is defined with an angle in degrees, default angle is 0: .
- The `labels `parameter must be an array with one label for each wedge: .
- The `explode parameter`, if specified, and not None, must be an array with one value for each wedge.
- `legend()` to display legends . 
- `shadow = True` to add shadow in pie chart


```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([35,25,25,20])
abels=["jan","feb","mar","apr"]
myexplode=[0.2,0.2,0,0]
plt.pie(y,labels=abels, startangle=90, explode=myexplode)
plt.legend(title="months")
plt.show()
```

# From Histogram and Boxplots several key insights can be analyzed:

`Histogram Analysis`:
 - `Distribuition Shape`: It shows how data is distributed across different values, revealing if data is symetric, skewed (left or right) or has mutiple peaks. 

 - `Central Tendency`: The peak of histogram indicates where most values are concentrated, helping to identify the mode.

 - `spread and variabiliy`: The width of histogram shows how spread out the data is -wider hisograms include higher variability. 

 - `Outliners and Gaps`: unusual gaps or isolated bars may indicate outliners or missing Data. 

 `BoxPlot Analysis`:
 - `Five number summary`: Boxplot displays the minimum, first quartile(Q1), median(Q2),  3rd quartile(Q3), and maximum giving a quick summary of data central tendency.

 - `Inquartile Range (IQR)`: The box itself represents the middle 50% of the data (Q1 to Q3), showing the Data variablity. 

 - `Outliners`: Points outside the whiskers are considered outliners, which can indicate unequal, the data is skewed (right or left).
 
 - `Skewness`: if median is not centered in the box or whiskers are unequal, the data is skewed. (right or left)

 - `comparison`: Boxplots are usually useful for compairing distribution across different groups or categories. 


# Common Diagram Types in Data Analysis

There are several other types of diagrams commonly used for data analysis, each providing unique insights depending on the data and the question being asked.

## Diagram Types and Their Uses

| Diagram Type         | Description                                                                 | Best For                                      |
|----------------------|-----------------------------------------------------------------------------|-----------------------------------------------|
| Bar Plot             | Compares values between different categories or groups                      | Categorical data, comparing counts, averages  |
| Line Plot            | Shows trends or changes over time or ordered variables                      | Visualizing continuous data, time series      |
| Scatter Plot         | Displays the relationship between two numeric variables                     | Identifying correlations, clusters, outliers  |
| Pie Chart            | Illustrates proportions or percentages of different categories              | Showing how parts make up a whole             |
| Area Plot            | Similar to a line plot, but the area under the line is filled               | Emphasizing cumulative change, trends         |
| Stacked Bar/Area Plot| Compares multiple groups or categories, with each group stacked on another  | Showing both individual and total values      |
| Table Visualization  | Presents raw data or summary statistics in a tabular format                 | Detailed inspection, summary statistics       |

## What Each Diagram Helps Analyze

| Diagram Type         | Analysis Insights                                                                 |
|----------------------|-----------------------------------------------------------------------------------|
| Bar/Stacked Bar      | Compare categories, identify largest/smallest groups, see group contributions     |
| Line/Area            | Track trends, spot patterns, see how values change over time                      |
| Scatter              | Detect relationships, correlations, or clusters between two variables             |
| Pie                  | Understand relative size of categories within a whole                             |
| Table                | Examine detailed data, check for missing values, review summary statistics        |
