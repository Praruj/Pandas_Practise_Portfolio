# Markers
We can use keyboard argument `marker` to emphasize each point with a specified marker.

```python
import matplotlib.pyplot as  plt
import numpy as np

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="*") #marker="o" for topping as 0. 
plt.show()
```

We can always find marker reference and use them according to our needs. 

Format strings: `fmt` 
`marker/Line/color` syntax can be used in shortcut string notation parameter to specify the marker. 

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,9,1,10])
plt.plot(ypoints, "o--b")
plt.show()
```

The marker value can be anything from Marker Reference and color value as well. 

```python
# Marker size: ms to set the size of markers. 

import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="o",ms=15)
plt.show()
```

```python
# Marker egde color by using mec, to set egde of marker.

import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([6,10,12,14])
plt.plot(ypoints,marker="o",ms=20,mec="g")
plt.show()
```

```python
# To set the markerface color, we use mfc to set color inside the edge of the markers.

import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([4,12,16,24])
plt.plot(ypoints,marker="*",ms=20,mfc="k")
plt.show()
```

```python
# Mark each point with a beautiful green color:
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
```

```python
# Matplotlib line: ls to change the style of plotted line. 

import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([12,16,18,20])
plt.plot(ypoints, linestyle="dashed") # dotted 
plt.show()
```

The `linestyle` can be written in shorter syntax:

- `linestyle` can be written as `ls`
- `dotted` can be written as `:`
- `dashed` can be written as `--`.

```python
# line colour:

import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([1,2,3,4,5])
plt.plot(ypoints, color="r")
plt.show()
```

```python
# line width
# line colour:

import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([1,2,3,4,5])
plt.plot(ypoints, color="r", linewidth="18")
plt.show()
```
