# Matplotlib core structure: Figures and Axes. 
`Figure` (the canvas) and  `Axes` (the actual plot area where data lives).  
A figure can hold multiple axes (subplots), and each Axes object is where we plot our data. 
<br>
“The greatest value of a picture is when it forces us to notice what we never expected to see.” – John Tukey

```python
from matplotlib import pyplot as plt
Age=[22,29,35,41,24,27]
height=[180,165,170,182,178,163]
plt.plot(Age,height)
plt.title("Age vs height Graph")
plt.xlabel("Age")
plt.ylabel("height")
plt.show()
```

```python
from matplotlib import pyplot as plt
Age=sorted([22,29,35,41,24,27])
height=sorted([180,165,170,182,178,163])
weight=sorted([59,76,82,80,87,67])

plt.plot(Age,height,"k--",label="height")
plt.plot(Age,weight,"b",label="weight")



plt.title("gym datasets with Age") 
plt.xlabel("Age")
plt.ylabel("height")

plt.legend()
plt.show()

#Graph was looking weird so here i sorted the age and height. 
# We can do format strings which have Marker, color and Line. 
#link: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
```

```python
# we can pass format as Arguments as well

from matplotlib import pyplot as plt
Age=sorted([22,29,35,41,24,27])
height=sorted([180,165,170,182,178,163])
weight=sorted([59,76,82,80,87,67])

plt.plot(Age,height,color="k",linestyle="--",label="height")
plt.plot(Age,weight,color="b",label="weight")

plt.title("gym datasets with Age") 
plt.xlabel("Age")
plt.ylabel("height")

plt.legend()
plt.show()
```

```python

from matplotlib import pyplot as plt
# we can pass format as Arguments as well

from matplotlib import pyplot as plt
Age=sorted([22,29,35,41,24,27])
height=sorted([180,165,170,182,178,163])
weight=sorted([59,76,82,80,87,67])

plt.plot(Age,height,color="k",linestyle="--",marker="o",label="height")
plt.plot(Age,weight,color="b",marker=".",label="weight")

plt.title("gym datasets with Age") 
plt.xlabel("Age")
plt.ylabel("height")

plt.legend()
plt.show()

# O will be large marker as we can see in the figure. 
```

```python
#  we can Hex values as well,  which have color paletee online; color=#444444. 

from matplotlib import pyplot as plt
# we can pass format as Arguments as well

from matplotlib import pyplot as plt
Age=sorted([22,29,35,41,24,27])
height=sorted([180,165,170,182,178,163])
weight=sorted([59,76,82,80,87,67])

plt.plot(Age,height,color="#444444",linestyle="--",marker="o",label="height")
plt.plot(Age,weight,color="#db5a04",marker=".",label="weight")

plt.title("gym datasets with Age") 
plt.xlabel("Age")
plt.ylabel("height")

plt.legend()
plt.show()
```
